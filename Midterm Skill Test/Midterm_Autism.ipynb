## (c) Fitting a Logistic Regression

### preprocessing 

```julia
data = dt[dt[!, :age].!="NA",:]
```

```julia
replace!(data.ethnicity, "NA" => "0" )
replace!(data.ethnicity, "White-European" => "1" )
replace!(data.ethnicity, "Latino" => "2" )
replace!(data.ethnicity, "Black" => "3" )
replace!(data.ethnicity, "Asian" => "4" )
replace!(data.ethnicity, "Middle Eastern " => "5" )
replace!(data.ethnicity, "Pasifika" => "6" )
replace!(data.ethnicity, "South Asian" => "7" )
replace!(data.ethnicity, "Hispanic" => "8" )
replace!(data.ethnicity, "Turkish" => "9" )
replace!(data.ethnicity, "Others" => "10" )
replace!(data.ethnicity, "others" => "11" )
```

```julia
data[!, :age] = parse.(Float64, data.age);
data[!, :ethnicity] = parse.(Float64, data.ethnicity);
```

```julia
data[!, :b_austim] = ifelse.(data.austim .== "yes", 1, 0);
data[!, :b_jundice] = ifelse.(data.jundice .== "yes", 1, 0); 
data[!, :b_used_app_before] = ifelse.(data.used_app_before .== "yes", 1, 0);
data[!, :b_gender] = ifelse.(data.gender.== "f", 1, 0);
data[!, :b_age] = ifelse.(data.age .>= mean(data.age), 1, 0);
```

```julia
describe(data)
```

```julia
select!(data, Not([:austim, :jundice, :age_desc, :Column1, :age, :gender, :used_app_before, :contry_of_res]))
```

```julia
sum(data.b_austim), length(data.b_austim) - sum(data.b_austim)
```

```julia
n, p = size(data)
```

```julia
Random.seed!(19247923)
ind = randperm(n);
train_df = data[1:500,:];
test_df = data[501:702,:];
```

```julia
describe(train_df)
```

```julia
fm = @formula(b_austim ~ A1_Score+ A2_Score+ A3_Score + A4_Score + A5_Score + A6_Score + A7_Score + A8_Score + A9_Score + A10_Score + b_jundice + b_used_app_before + relation + b_gender + ethnicity + b_age)
```

```julia
logit = glm(fm, train_df, Binomial(), LogitLink())
```

```julia
deviance(logit)
```

## (e) Accuracy & ROC

```julia
tpred = predict(logit, test_df);
target = test_df.b_austim;
```

```julia
using RCall
```

```julia
R"library(ROCR)"
```

```julia
@rput tpred
@rput target
```

### threshold 0.5

```julia
using EvalMetrics

th = 0.5
tpred_demo =ifelse.(tpred .>= th , 1 , 0);
accuracy(target, tpred_demo)
```

```julia
R"pred= prediction(tpred,target)"
R""" perf= performance(pred, "tpr", "fpr") """
```

```julia
R"plot(perf,colorize=TRUE)"
```

```julia
thres = 0.6
tpredbin=ifelse.(tpred .>= thres , 1 , 0);
```

```julia
using EvalMetrics

accuracy(target, tpredbin)
```

## (f) Confusion Matrix

```julia
using MLBase

confusion_matrix = MLBase.roc(target, tpredbin)
```

- p --> number of actual positive
- n --> number of actual negative

```julia
TP = confusion_matrix.tp
TN = confusion_matrix.tn
FP = confusion_matrix.fp
FN = confusion_matrix.fn;
```

```julia
acrcy = (TP+TN) / (TP+TN+FP+FN)
```

```julia
sensitivity = (TP) / (TP + FN)
```

```julia
specificity = (TN) / (FP + TN)
```

```julia
ppv = (TP) / (TP + FP)
```

```julia
npv = (TN) / (TN + FN)
```

### The probability of the model yeils a correct resutl is around 85% The probability that a positve test will truly have a disease is 50% & the negative test truly be disease free is 85% . This is really important (npv) to interpret the model is truly worked or not in this context. The probability that a true negative will test negative is also very high about 99%, that is also a good indication of the model.
