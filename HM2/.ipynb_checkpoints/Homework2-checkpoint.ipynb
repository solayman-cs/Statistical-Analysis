# 1. Test Statistic

### T-Test

```julia
using HypothesisTests
x=diabetes[!,:Pregnancies]
t1 = OneSampleTTest(x, 8)
```

```julia
pvalue(t1, tail = :both)
```

```julia
pvalue(t1, tail = :left)
```

```julia
pvalue(t1, tail = :right)
```

-  We can't conclude that the mean of the population from which the sample was drawn is greater than 8 as we haven't enough evidence to claim that. It is also verified by the  Hypothesis Testing *(T-test)* 

# 2. Bootstraping to find the sampling distribution

```julia
B = 500
x_bar = mean(x)
BMean = zeros(B)
tobs=zeros(B)
atobs=x_bar/(std(x)/sqrt(n))

for i in 1:B
    data = sample(x, n, replace=true, ordered=false)
    BMean[i] = mean(data)
    tobs[i] = mean(data)/(std(data)/sqrt(n))
end
```

```julia
# confidence interval
cp = percentile(BMean, [2.5,97.5])

# p-value
mean(BMean .>= x_bar)
```

```julia
mean(tobs .>= atobs)
```

```julia
Gadfly.plot(x=BMean, Geom.histogram)
```

```julia
Gadfly.plot(x=tobs,Geom.histogram)
```

```julia
 StatsPlots.boxplot(BMean)
```

```julia
 StatsPlots.boxplot(tobs)
```

# 3. Claim about the Proportion

```julia
out = diabetes[diabetes.Outcome .== 1, :Outcome] 
typeof(out)
```

```julia
samp_prop = length(out)/n
```

## Z-test

```julia
assm_prop = 0.5
nume = samp_prop - assm_prop
deno = assm_prop*(1-assm_prop)
Z_test = nume / sqrt(deno/length(out))
```

- By observing the *Z-Test*, we have enough evidence to claim that the proportion of diabetes women in the population is different from 0.5

# 4. Repeat Procedure for the "Outcome" Variable

```julia
Outcome = diabetes[!,:Outcome]

B = 500
obs = zeros(B)
sel_obs = zeros(B)
prop = zeros(B)

for i in 1:B
    obs = sample(Outcome, n, replace=true, ordered=false) 
    sel_obs = obs[obs .== 1] 
    prop[i] = length(sel_obs) / n       
end

```

```julia
prop
```

```julia
Gadfly.plot(x=prop, Geom.histogram)
```

```julia
 StatsPlots.boxplot(prop)
```

```julia
assm_prop = 0.5
nume = mean(prop) - assm_prop
deno = assm_prop*(1-assm_prop)
Z_test = nume / sqrt(deno/n)
```

```julia
mean(prop)
# confidence interval
cp=percentile(prop, [2.5,97.5])
# p-value
mean(prop .>= samp_prop)
```
